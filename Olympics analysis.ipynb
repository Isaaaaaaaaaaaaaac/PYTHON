```python
#Top countries participating
top_countries = df.Team.value_counts().sort_values(ascending = False).head(10)
top_countries
```

```python
top_countries = df['Team'].value_counts().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='Set2')
plt.title('Top 10 Countries Participating')
plt.xlabel('Country')
plt.ylabel('Participation Count')
plt.show()
```

```python
#Age distribution of participants
plt.figure(figsize=(12,6))
plt.title('Age distribution of athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(df.Age ,bins = np.arange(10,80,2) ,color = 'orange', edgecolor = 'white' )
plt.show()
```

```python
# Total numbers of female athletes in each Olympics
female_participants = df[(df.Sex == 'F') & (df.Season == 'Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail(10)
```

```python
womenolympics = df[(df.Sex == 'F') & (df.Season == 'Summer')]
```

```python
sns.set(style = 'darkgrid')
plt.figure(figsize = (12,5))
sns.countplot(x = 'Year', data = womenolympics, palette = 'Spectral')
plt.title('Women participantion')
```

```python
part = womenolympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize = (12,5))
part.loc[ :,'F'].plot()
plt.title('Women participatation over time')
```

```python
#Gold medal athletes
gold_medal = df[(df.Medal == 'Gold')]
gold_medal.head()
```

```python
#Take only the values that are different from the NaN
gold_medal = gold_medal[np.isfinite(gold_medal['Age'])]
```

```python
#Gold beyond 60
gold_medal['ID'][gold_medal['Age']>60].count()
```

```python
sporting_event =gold_medal['Sport'][gold_medal['Age']>60]
sporting_event
```

```python
plt.figure(figsize = (12,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medal for athletes over 60 years')
```

```python
total_goldmedals = gold_medal['Team'].value_counts().reset_index(name = 'Medal').head()
```

```python
total_goldmedals = gold_medal['Team'].value_counts().reset_index(name = 'Medal').head(6)
a = sns.catplot(x = 'index', y='Medal', data =total_goldmedals, height = 5, kind = 'bar', palette ='rocket') 
a.despine(left = True)
a.set_xlabels('Countries')
a.set_ylabels('Numbers of Medal')
plt.title('Gold Medal Per Country')
```
