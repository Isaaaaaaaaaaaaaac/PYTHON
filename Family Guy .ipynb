```python
# draw regplot
import seaborn as sns
sns.regplot(x = 'No. of Episode (Overall)',
            y = 'IMDb Rating',
            data = df,
           scatter_kws={"color": "black"}, line_kws={"color": "red"})
```

```python
import matplotlib.pyplot as plt
df["IMDb Rating"].plot.hist()
plt.title("Family Guy IMDb Rating")
plt.show
```

```python
df.groupby('IMDb Rating').count()['U.S. Viewers (Millions)'].plot()
plt.title('Family Guy Rating/Number of US Viewers', fontsize=20);
plt.tight_layout()
```

```python
sns.lmplot(x="No. of Episode (Overall)", y="IMDb Rating", hue="Season", data=df, height=6, aspect=2, ci=None)
plt.xlabel("Episode")
plt.ylabel("IMDb Rating")
plt.title("IMDb Ratings for Family Guy Episodes");
```

```python
sns.lmplot(x="No. of Episode (Overall)", y="Running Time (Minutes)", hue="Season", data=df, height=6, aspect=2, ci=None)
plt.xlabel("Episode")
plt.ylabel("Running time in Minutes")
plt.title("Running Time for Family Guy episodes")
plt.ticklabel_format(style='plain', axis='y');
```
