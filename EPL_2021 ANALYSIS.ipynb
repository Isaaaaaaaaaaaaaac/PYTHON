```python
#liberies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df = pd.read_csv('C:/pythonprojects/EPL_20_21.csv')
epl_df.head()
```

```python
#data info
epl_df.info()
```

```python

epl_df.describe()
```

```python
#total null
epl_df.isnull().sum()
```

```python
#two new columns added 'MinsPerMatch' 'GoalPerMatch'
epl_df['MinsPerMatch'] = epl_df['Mins'] / epl_df['Matches'].astype(int)
epl_df['GoalPerMatch'] = epl_df['Goals'] / epl_df['Matches'].astype(float)
```

```python
#total penalty attempted
sum(epl_df['Penalty_Attempted'])
```

```python
plt.figure(figsize =( 15 ,6))
pl_not_scored = sum(epl_df['Penalty_Attempted']) - sum(epl_df['Penalty_Goals'])
pl_Goals = sum(epl_df['Penalty_Goals'])
data = [pl_not_scored,pl_Goals]
lables = ['Penalty missed' , 'Penalty scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = lables , colors = color , autopct = '%.0f%%')
plt.show()
```

```python
# Most players from which countries
nationalty = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationalty.head(10).plot(kind = 'bar', figsize = (10,5) , color = sns.color_palette('magma'))
```

```python
# club with maximum playes in their squad
epl_df['Club'].value_counts().nlargest(5).plot(kind = 'bar' , figsize = (12,5), color = sns.color_palette('magma'))
```

```python
# club with least playes in their squad
epl_df['Club'].value_counts().nsmallest(5).plot(kind = 'bar' , figsize = (12,5), color = sns.color_palette('magma'))
```

```python
#players based on age group
under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
above30 = epl_df[epl_df['Age'] > 30]

x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
my_labels = ['<=20', '>20 & <=25', '>25 & <=30', '>30']
plt.title('Total player with Age', fontsize=20)
plt.pie(x, labels=my_labels, autopct='%.0f%%')
plt.show()                  
```

```python
#Average age player in each club

plt.figure(figsize = (12,8))
sns.boxplot(x = 'Club' , y = 'Age' , data = epl_df)
plt.xticks(rotation = 90)
```

```python
num_player = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
#total assists from each club
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=Assists_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=20)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20, 18)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
plt.show()
```

```python
#total assits from each club
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=Assists_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=20)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20, 18)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
plt.show()
```

```python
#total goals from each clubs
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=Goals_by_clubs.sort_values(by='Goals'), palette='rocket')
ax.set_xlabel('Club', fontsize=20)
ax.set_ylabel('Goals', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20, 18)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
plt.show()
```
